# Model Evaluation Phase

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Read CSV file

```python
ModelResults = pd.read_csv('CNNResults.csv')
loss = ModelResults['loss']
val_loss=ModelResults['val_loss']
acc = ModelResults['accuracy']
val_acc=ModelResults['val_accuracy']
```

```python
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show
```

```python
plt.plot(epochs, acc, 'bo', label='Training Accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

# Deployment Phase

## Load model and Image For prediction

```python
from keras.models import load_model
from keras.preprocessing import image
from keras.utils import load_img
from keras.utils import img_to_array
import numpy as np
import matplotlib.pyplot as plt

# Load the saved model
model = load_model('big_cat.h5')

# Load the input image
img = load_img('prediction_images/cheetah5.jpg',target_size=(150,150))
img_array = img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array /= 255.0  # Normalize the image

# Make predictions
predictions = model.predict(img_array)

# Interpret predictions
class_index = np.argmax(predictions)
class_labels = ['cheetah', 'leopard', 'lion', 'tiger']
predicted_class = class_labels[class_index]

# Display the result
print(f'The predicted class is: {predicted_class}')

# Display the image
plt.imshow(img)
plt.show()
```
